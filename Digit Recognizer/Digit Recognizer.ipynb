```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.layers import Input, Dense
from keras.models import Model
from keras import regularizers
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
sample = pd.read_csv('data/sample_submission.csv')
```

```python
comp_dim = 64

inp = Input(shape=(784,))
encoded = Dense(comp_dim, activation='relu')(inp)
decoded = Dense(784, activation='sigmoid')(encoded)

autoencoder = Model(inp, decoded)
encoder = Model(inp,encoded)
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python
X = train.drop(columns='label').copy()
X = X.astype('float32') / 255
X_train, X_test, y_train, y_test = train_test_split(X, X, test_size=0.2, random_state=42)
```

```python
autoencoder.fit(X_train,X_train,epochs=200,batch_size=256,shuffle=True,validation_data=(X_test, X_test))
```

```python
encoded_data = encoder.predict((X))
cols = []
for i in range(comp_dim):
    cols.append('columns'+str(i))
df = pd.DataFrame(data = encoded_data,columns=cols)
```

```python
from keras.utils.np_utils import to_categorical
y = train['label']
y= to_categorical(y)
cols = []
for i in range(10):
    cols.append(i)
y = pd.DataFrame(data=y,columns=cols)
```

```python
from keras.models import Sequential
classif = Sequential()
classif.add(Dense(256, activation='relu',input_shape=(64,)))
classif.add(Dense(10,activation='softmax'))
from keras.optimizers import RMSprop
classif.compile(optimizer=RMSprop(lr=0.001),
 loss='categorical_crossentropy',
 metrics=['accuracy'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
```

```python
classif.fit(X_train,y_train,epochs=200,batch_size=256,shuffle=True,validation_data=(X_test, y_test))
```

```python
test = test/255
encoded_data = encoder.predict(test)
cols = []
for i in range(comp_dim):
    cols.append('columns'+str(i))
df = pd.DataFrame(data = encoded_data,columns=cols)
```

```python
predictions = classif.predict_classes(df)
```
